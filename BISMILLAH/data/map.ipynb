```python
import pandas as pd
```

```python
df = pd.read_csv('Air_Traffic_Passenger_Statistics.csv')
```

```python
df.head()
```

```python
# Convert the date column to datetime type
df['Activity Period'] = pd.to_datetime(df['Activity Period'], format='%Y%m')

# Extract year and month from the datetime column
df['year'] = df['Activity Period'].dt.year
df['month'] = df['Activity Period'].dt.month

# Group the data by region, year, and month and calculate the sum of passenger counts
grouped_df = df.groupby(['GEO Region', 'year'])['Passenger Count'].sum().reset_index(name='passenger_count')

# Display the resulting DataFrame
print(grouped_df)
```

```python
grouped_df.to_csv('map_data.csv',index=False)
```
