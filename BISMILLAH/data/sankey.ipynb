```python
import pandas as pd
df = pd.read_csv('Air_Traffic_Passenger_Statistics.csv')
df.head()
```

```python
import pandas as pd

# Assuming you have a DataFrame called `df` with the necessary data

# Group the data by 'GEO Region' and calculate the sum of 'Passenger Count'
grouped_df = df.groupby('GEO Region')['Passenger Count'].sum().reset_index()

# Create a new DataFrame to store the Sankey diagram data
sankey_data = pd.DataFrame(columns=['source', 'target', 'value'])

# Set the source node as "San Francisco Airport"
sankey_data['source'] = ['San Francisco Airport']*len(grouped_df)

# Fill the target column with the regions
sankey_data['target'] = grouped_df['GEO Region']

# Fill the value column with the passenger counts
sankey_data['value'] = grouped_df['Passenger Count']
```
